*`Biblioteas Usadas`*

```python
from faker import Faker
from random import *
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud
from wordcloud.wordcloud import STOPWORDS
from num2words import num2words
```

```python
fk = Faker()
x = 30
lista=['coluna1']
for valores in range(x):
  s = fk.name()
  lista.append(s)
 
lista
```

*`Cria√ß√£o do arquivo txt, para nomes`*

```python
with open('arquivo1.txt','w') as arquivo:
  for valor in lista :
    arquivo.write(str(valor)+'\n')
```

*`Gerador de Numeros`*

```python
N=10
```

```python
lista_a = [randint(0,10) for s in range(N)]
lista_b = lista_a
```

```python
lista_b
```

`Arquivo para numeros`

```python
with open('arquivo 2.txt','w') as pontuacoes:
  pontuacoes.write('coluna1\n')
  for valor in lista_b:
    pontuacoes.write(str(valor)+'\n')
# numeros por extenso
with open('nm extenso.txt','w') as nm:
  nm.write('indice\n')
  for numero in lista_b:
    tranformar_n = num2words(numero)
    nm.write(tranformar_n+ '\n')
```

`Cria√ß√£o dos histograma`

```python
plt.hist(lista_b,label ='Pontua√ß√µes')
plt.legend()
plt.show()
```

`Nuvem De palavras`

```python
df = pd.read_table("/content/nm extenso.txt")
sumary = df.dropna(subset=['indice'],axis=0)['indice']
```

```python
all_sumary = " ".join(s for s in sumary )
```

```python
print(len(all_sumary))
```

```python
stopwords = set(STOPWORDS)
```

```python
word = WordCloud(stopwords=stopwords,
                 background_color="black",
                 width=1600,height=800).generate(all_sumary)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(word,interpolation='bilinear')
ax.set_axis_off()
```
